```python
class kmeans: 
    def __init__(self, num_clusters):
        self.K = num_clusters # cluster number
        self.max_iterations = 100 # max iteration. don't want to run inf time
        self.num_examples, self.num_features = x_train.shape # num of examples, num of features
        self.plot_figure = True # plot figure

    # randomly initialize centroids
    def initialize_random_centroids(self, x_train):
        centroids = np.zeros((self.K, self.num_features)) # row , column full with zero 
        for k in range(self.K): # iterations of 
            centroid = x_train[np.random.choice(range(self.num_examples))] # random centroids
            centroids[k] = centroid
        return centroids # return random centroids

    # create cluster Function
    def create_cluster(self, x_train, centroids):
        clusters = [[] for _ in range(self.K)]
        for point_idx, point in enumerate(x_train):
            closest_centroid = np.argmin(
                np.sqrt(np.sum((point-centroids)**2, axis=1))
            ) # closest centroid using euler distance equation(calculate distance of every point from centroid)
            clusters[closest_centroid].append(point_idx)
        return clusters  

    # new centroids
    def calculate_new_centroids(self, cluster, x_train):
        centroids = np.zeros((self.K, self.num_features)) # row , column full with zero
        for idx, cluster in enumerate(cluster):
            new_centroid = np.mean(x_train[cluster], axis=0) # find the value for new centroids
            centroids[idx] = new_centroid
        return centroids  

    # prediction
    def predict_cluster(self, clusters, x_train):
        y_pred = np.zeros(self.num_examples) # row1 fillup with zero
        for cluster_idx, cluster in enumerate(clusters):
            for sample_idx in cluster:
                y_pred[sample_idx] = cluster_idx
        return y_pred
    
    # Fit the data : 
    def fit(self, x_train):
        centroids = self.initialize_random_centroids(x_train) # initialize random centroids
        for _ in range(self.max_iterations):
            clusters = self.create_cluster(x_train, centroids) # create cluster
            previous_centroids = centroids
            centroids = self.calculate_new_centroids(clusters, x_train) # calculate new centroids
            diff = centroids - previous_centroids # calculate difference
            if not diff.any():
                break
        y_pred = self.predict_cluster(clusters, x_train) # predict function
        if self.plot_figure: # if true
            self.plot_fig(x_train, y_pred) # plot function 
        return y_pred


# Initiate the model :
kmeans = KMeans(n_clusters=2, max_iter=100)


# Train the model :

kmeans.fit(x_train)

# Tester le model : 

kmeans.labels_
```
